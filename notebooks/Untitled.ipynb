#### Where do we begin?

One good place to be would be with the `NaaNs`. If there are any columns with a large about of data missing (i.e., 50% of more) -- dropping these columns from the dataset could be a good place to start.

```python
# Let's look at 100 columns with the most NaaNs
bos.isna().sum().sort_values(ascending = False)[:100]
```

100 columns in the dataset have **75% or more** missing data! The data in these columns may have been valuable, but I wouldn't feel comfortable filling `NaaNs` with any values (i.e., mean value of the column) with so much data missing. Let's drop these columns from the dataset.

```python
[column for column in bos.columns if bos[column].isna().sum() >= 105]
```

```python
# Drop columns with over 105 NaaNs from dataset
bos = bos.drop(columns = [column for column in bos.columns if bos[column].isna().sum() >= 105])

# Check work
bos.isna().sum().sort_values(ascending = False)
```

Even after dropping 100 columns from our dataset that were riddled with `NaaNs`, we still have 151 columns, some of which still have over 50% `NaaNs` in them (see above). Let's take a look at ant columns with over `50%` of their values missing from the dataset.

```python
[column for column in bos.columns if bos[column].isna().sum() >= bos.shape[0] / 2]
```

```python
# Drop columns with over 50% NaaNs from dataset
bos = bos.drop(columns = [column for column in bos.columns if bos[column].isna().sum() >= bos.shape[0] / 2])

# Check work
bos.isna().sum().sort_values(ascending = False)
```

### Currency Formatting

A few columns report data in the form of currencies. These objects should be housed as `floats` but are instead are stored as `objects` with several symbols and punctuation marks embedded in each cell. The columns are fiull of `NaNs`, and in order to fill those empty cells with a useful value, we're going to have to remove to symbols from each cell of each column and convert the `object` to a float.

The below funciton has been created to do just that.

```python
# Define function to change values from objects to floats
def object_currency_to_float(column):
    
    # Create empty list
    new_column = []
    
    # Loop through each cell in the inputted column
    for cell in bos[column]:
        
        # Check to see if each cell is a string
        if type(cell) == str:
            
            cell = cell.replace('$', '')
    
            cell = cell.replace('-', '0')
            
            cell = cell.replace(',', '')
                
            cell = float(cell)
            
            new_column.append(cell)
            
        else:
            
            new_column.append(cell)
    
    return new_column
    
```

**BPS_Electric_Bill**

```python
# Call function and set equal to the Electric Bill column
bos['BPS_Electric_Bill'] = object_currency_to_float('BPS_Electric_Bill')
```

```python
# Fill NaaNs in this column with the average value of Electric Bills that we have available 
#\n in the dataset
bos['BPS_Electric_Bill'].fillna(bos['BPS_Electric_Bill'].mean(), inplace = True)
```
